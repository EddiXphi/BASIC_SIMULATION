## **1. Calcule la probabilidad de obtener exactamente 3 éxitos en 8 ensayos, con una probabilidad de éxito de 0.6.**


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

#Parámetros de la distribución binomial
n = 8  #número total de ensayos
p = 0.6  #probabilidad de éxito en cada ensayo

#Generar la PMF (función de masa de probabilidad)
#y la CDF (función de distribución acumulada)
#para todos los posibles números de éxitos
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
cdf = binom.cdf(x, n, p)

#Generar el histograma de variables aleatorias de la distribución binomial
random_variables = binom.rvs(n, p, size=1000)

# Dibujar el PMF
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.bar(x, pmf, color='blue', alpha=0.7)
plt.title("Función de Masa de Probabilidad (PMF)")
plt.xlabel("Número de Éxitos")
plt.ylabel("Probabilidad")

# Dibujar el CDF
plt.subplot(1, 3, 2)
plt.step(x, cdf, where='mid', color='green')
plt.title("Función de Distribución Acumulada (CDF)")
plt.xlabel("Número de Éxitos")
plt.ylabel("Probabilidad Acumulada")



# Dibujar el Histograma
plt.subplot(1, 3, 3)

plt.hist(random_variables, bins=range(n+2), align='left', color='orange', alpha=0.7)
plt.title("Histograma de Variables Aleatorias")
plt.xlabel("Número de Éxitos")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()

mean, var, skew, kurt = binom.stats(n,p,moments = "msvk")

print("Media: ", mean)

print("Varianza: ", var)

print("Sesgo: ", skew)

print("Curtosis: ", kurt)
```

## **2. ¿Cuál es la probabilidad de que ocurran 5 o menos eventos en un intervalo de tiempo, dado que la tasa media de eventos es 3 por intervalo?**



```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la distribución de Poisson
lambda_ = 3  # tasa media de eventos por intervalo

# Generar la PMF y la CDF para el rango de eventos
x = np.arange(0, 15)
pmf = poisson.pmf(x, lambda_)
cdf = poisson.cdf(x, lambda_)

# Calcular la probabilidad de 5 o menos eventos
probabilidad = poisson.cdf(5, lambda_)
print(f"La probabilidad de que ocurran 5 o menos eventos es: {probabilidad:.4f}")

# Generar un histograma de variables aleatorias de la distribución de Poisson
random_variables = poisson.rvs(lambda_, size=1000)

# Dibujar el PMF
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.bar(x, pmf, color='purple', alpha=0.7)
plt.title("Función de Masa de Probabilidad (PMF)")
plt.xlabel("Número de Eventos")
plt.ylabel("Probabilidad")

# Dibujar el CDF
plt.subplot(1, 3, 2)
plt.step(x, cdf, where='mid', color='red')
plt.title("Función de Distribución Acumulada (CDF)")
plt.xlabel("Número de Eventos")
plt.ylabel("Probabilidad Acumulada")

# Dibujar el Histograma
plt.subplot(1, 3, 3)
plt.hist(random_variables, bins=range(16), align='left', color='green', alpha=0.7)
plt.title("Histograma de Variables Aleatorias")
plt.xlabel("Número de Eventos")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()


mean, var, skew, kurt = poisson.stats(lambda_,moments = "msvk")

print("Media: ", mean)

print("Varianza: ", var)

print("Sesgo: ", skew)

print("Curtosis: ", kurt)
```


## **3. En un lote de 20 artículos, hay 5 defectuosos. Si seleccionamos al azar 4 artículos, ¿cuál es la probabilidad de que exactamente 2 sean defectuosos?**


```python
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la distribución hipergeométrica
M = 20  # tamaño total de la población (total de artículos)
n = 5   # total de objetos de tipo deseado en la población (artículos defectuosos)
N = 4   # tamaño de la muestra extraída de la población

# Calcular la probabilidad de que exactamente 2 de los 4 artículos sean defectuosos
k = 2  # número de artículos defectuosos deseados en la muestra
probabilidad = hypergeom.pmf(k, M, n, N)
print(f"La probabilidad de que exactamente 2 de los 4 artículos sean defectuosos es: {probabilidad:.4f}")

# Generar la PMF y la CDF para el rango de artículos defectuosos posibles en la muestra
x = np.arange(0, N+1)
pmf = hypergeom.pmf(x, M, n, N)
cdf = hypergeom.cdf(x, M, n, N)

# Generar un histograma de variables aleatorias de la distribución hipergeométrica
random_variables = hypergeom.rvs(M, n, N, size=1000)

# Dibujar el PMF
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.bar(x, pmf, color='brown', alpha=0.7)
plt.title("Función de Masa de Probabilidad (PMF)")
plt.xlabel("Número de Artículos Defectuosos")
plt.ylabel("Probabilidad")

# Dibujar el CDF
plt.subplot(1, 3, 2)
plt.step(x, cdf, where='mid', color='blue')
plt.title("Función de Distribución Acumulada (CDF)")
plt.xlabel("Número de Artículos Defectuosos")
plt.ylabel("Probabilidad Acumulada")


# Dibujar el Histograma
plt.subplot(1, 3, 3)
plt.hist(random_variables, bins=range(N+2), align='left', color='grey', alpha=0.7)
plt.title("Histograma de Variables Aleatorias")
plt.xlabel("Número de Artículos Defectuosos")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()


rv= hypergeom(M, k, N)
mean, var, skew, kurt =rv.stats(moments= "msvk")
print("Media: ", mean)
print("Varianza: ", var)
print("Sesgo: ", skew)
print("Curtosis: ", kurt)
```



## 4. **Calcule la probabilidad de obtener un fracaso (0) en un ensayo, con una probabilidad de éxito del 0.75.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Parámetros del problema
probabilidad_exito = 0.75

# Valores posibles para el resultado del ensayo (0 para fracaso, 1 para éxito)
#resultados = np.array([0, 1])
resultados = bernoulli.rvs(probabilidad_exito, size=1)


# Función de masa de probabilidad (PMF)
pmf = bernoulli.pmf(resultados, probabilidad_exito)

# Función de densidad de probabilidad (PDF)
pdf = bernoulli.pmf(resultados, probabilidad_exito)

# Función de distribución acumulada (CDF)
cdf = bernoulli.cdf(resultados, probabilidad_exito)

# Histograma
# Puedes usar plt.bar para un histograma discreto
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)

plt.bar(resultados, pmf, align='center', alpha=0.7, color='blue', label='PMF')
plt.xlabel('Resultado del ensayo')
plt.ylabel('Probabilidad')
plt.title('Función de Masa de Probabilidad (PMF) de Bernoulli')
plt.xticks(resultados)

plt.subplot(1, 3, 2)
# También puedes trazar la función de distribución acumulada (CDF)
plt.step(resultados, cdf, color='green', where='post', label='CDF')
plt.xlabel('Resultado del ensayo')
plt.ylabel('Probabilidad acumulativa')
plt.title('Función de Distribución Acumulativa (CDF) de Bernoulli')
plt.xticks(resultados)

# Dibujar el Histograma
plt.subplot(1, 3, 3)
plt.hist(resultados, bins=range(2), align='left', color='grey', alpha=0.7)
plt.title("Histograma de Variables Aleatorias")
plt.xlabel("Número de Artículos Defectuosos")
plt.ylabel("Frecuencia")


plt.tight_layout()
plt.show()

mean, var, skew, kurt = bernoulli.stats(probabilidad_exito,moments = "msvk")

print("Media: ", mean)

print("Varianza: ", var)

print("Sesgo: ", skew)

print("Curtosis: ", kurt)
```


## **9. Un test diagnóstico para una enfermedad tiene una precisión del 95%. Si se aplica a un paciente que se sabe está enfermo, ¿cuál es la probabilidad de que el test arroje un resultado positivo (correcto)?**


```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de la distribución de Bernoulli
p = 0.95  # probabilidad de un resultado positivo correcto

# La PMF y la CDF
x = [0, 1]  # 0 para resultado negativo, 1 para resultado positivo
pmf = bernoulli.pmf(x, p)
cdf = bernoulli.cdf(x, p)

# Generar un histograma de variables aleatorias
random_variables = bernoulli.rvs(p, size=1000)

# Mostrar la probabilidad de un resultado positivo correcto
print(f"La probabilidad de un resultado positivo correcto es: {p:.2f}")

# Dibujar el PMF
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.bar(x, pmf, color='blue', alpha=0.7)
plt.title("Función de Masa de Probabilidad (PMF)")
plt.xticks(x, ['Negativo', 'Positivo'])
plt.ylabel("Probabilidad")

# Dibujar el CDF
plt.subplot(1, 3, 2)
plt.step(x, cdf, where='mid', color='green')
plt.title("Función de Distribución Acumulada (CDF)")
plt.xticks(x, ['Negativo', 'Positivo'])
plt.ylabel("Probabilidad Acumulada")

# Dibujar el Histograma

plt.subplot(1, 3, 3)

plt.hist(random_variables, bins=np.arange(-0.5, 2, 1), align='mid', color='red', alpha=0.7)
plt.title("Histograma de Variables Aleatorias")
plt.xticks([0, 1], ['Negativo', 'Positivo'])
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```
